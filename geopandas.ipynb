```python
import numpy as np
import pandas as pd
```

```python
import geopandas as gpd
from pyogrio import set_gdal_config_options
set_gdal_config_options({
    'SHAPE_RESTORE_SHX': 'YES',
})
```

```python
fp = r'Downloads\india-polygon.shp'
map_df = gpd.read_file(fp)
map_df_copy = gpd.read_file(fp)
map_df.head()
```

```python
map_df['st_nm']
```

```python
df = pd.read_csv('Downloads/combined_data_biometric.csv')
```

```python
total_peope_all_days = df.groupby(['state'])['total_biometric'].sum()
count = df.groupby(['state'])['total_biometric'].count()
```

```python
avg_people_all_day = total_peope_all_days/count
```

```python
merged = map_df.set_index('st_nm').join(avg_people_all_day)
```

```python
merged
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, figsize=(10, 10))
ax.axis('off')
ax.set_title('Number of Biometrics Updation state-wise', fontdict={'fontsize': '20', 'fontweight' : '10'})
# Plot the figure
merged.plot(column='total_biometric',cmap='YlOrRd', linewidth=0.8, ax=ax, edgecolor='0',legend=True,markersize=[39.739192, -104.990337], legend_kwds={'label': "Average enrollment in a day"})
```
